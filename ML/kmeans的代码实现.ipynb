```python
# 定义一个类：
# 其实很好理解
class KMeans1:
    # 我们要得到数据，传他的簇有多少个
    def __init__(self,data,num_clusters):
        # 【这里只需要把数据传进来就可以了】
        self.data = data
        self.num_clusters = num_clusters
    def train(self,max_iterations): # 
        # 传进去最大迭代次数：就是一直循环找到最佳点，要更新多少次，设置的次数
        
        # 先选择到当前的中间点
        # 中心点的初始化：
        # 往里面传数据，传他中心点的个数
        # 这个质心初始化方法等下还要初始化
       
    
        centroids = KMeans.centroids_init(self.data,self.num_clusters)
        # 这个是是第一步，先拿到我的中心点
        
        # 【下面是：】我的训练模块
        num_examples = self.data.shape[0]
        # 先把里面的值全部设为1
        # 【里面可以放下全部点，先把每个值设为1】
        # 等下放离每个点最近的质心
        closest_centroids_ids = np.empty((num_examples,1))
        for _ in range(max_iterations):
            # 先得到距离：
            # 下面这个就是【创建一个新函数：还没初始化】
            # 要算和各个质心的距离，所以要传进去各个质心,并找到最近哪一个
            closest_centroids_ids = kMeans.centroids_find_closest(self.data,centroids)
            
            
            # 我现在要进行中心位置更新了
            # 后面那个离谁最近，就取哪个
            centroids = KMeans.centroids_compute(self.data,closest_centroids_ids,num_clusters)
            # 上面这句在for循环里面就是不断地更新，不断调整
        
        # 【可以返回结果了】
        return centroids,closest_centroids_ids    
        
        
        
    @ staticmethod # 用这个就是为了我直接用函数名调用即可
    def centroids_init(self,data,num_clusters):
        # 我要一个随机的数据
        # 【看一共有多少数据】：
        num_examples = data.shape[0]
        # 在数据里面随机选择一些点【个数就是设置的簇的个数】作为最开始的质心：
        # 先是【洗牌】的操作，【洗牌然后再抽牌】
        random_ids = np.random.permutation(num_examples)
        # 左边的就是取随机到的点的顺序的前几个作为我的最开始的质心
        # 后面的冒号表示取到所有特征，要拿四维的数据
        centroids = data[random_ids[:num_clusters],:]
        # 上面就是初始化中心点
        return centroids
    
    
    @staticmethod 
    # 填坑，定义上面的centroids_find_closest(self.data,centroids)函数
    # 要算和各个质心的距离，所以要传进去各个质心,并找到最近哪一个
    def centroids_find_closest(self,data,centroids):
        # 开始计算：
        # 先把数据摆出来：
        num_examples = self.data.shape[0]
        # 下面是指中心点的个数
        num_centroids = centroids.shape[0]
        # 对存储各个点的最近的距离的质心的数组进行初始化：【就只存最近的】
        closest_centroids_ids = np.zeros((num_examples,1))
        
        # 要来个双层循环：求各个样本点到各个簇的距离：
            # 第一个先摆出来全部数据
        
        
   
        for examples_index in range(num_examples):
            # 第二个摆出来k个质心
            
            # 我要定义距离数组：
            distance = np.zeros(num_centroids)
            for centroid_index in range(num_centroids):
                # 第个冒号是指【这个数据的每一维度我都要】
                distance_diff = data[examples_index,:] - centroids[centroid_index,:]
                
                # ！！！！【我去，这一步很妙！】前面这个就是直接点与点，特征与特征对齐了，直接求出了每个特征和质心的差值
                # 后面我用欧几里得距离公式直接平方累加就行了
                distance[centroid_index] = np.sum(distance_diff**2)
            closest_centroids_ids[examples_index] = np.argmin[distance]
            
            # 【这个应该是表示找到最近的点的意思！！！】
            
            
        # 一个一个点算，这样就算完了
        
        # 【那个different不用建立二维的！！！是因为每次我都会求最小值】
            return closest_centroids_ids
         # 得到每一个样本点到对应中心的距离了
            
    @ staticmethod       
    def centroids_compute(self,data,closest_centroids_ids,num_clusters):
        # 这个就是进行均值的更新就可以了
        # 【要在各个维度都求一遍均值】
        num_features = data.shape[1]
        # 对每一个维度我都要计算
        centroids = np.zeros((num_clusters,num_features))
        for centroid_id in range(num_clusters):
            closest_ids = closest_centroids_ids
            
            # 这个就是把他搞到一维数据去
            
            centroids[closest_ids] = np.means(data[closest_ids.flatten(),:],axis = 0)
            # 先得到最近的索引
        return centroids 
    # 现在就填充完了，可以传回去了
```

```python
# 创建一个属于我自己的kmeans算法：
class MyKMeans:
    # 先想想我要啥值？
    # 首先我要【质心的数量（分堆的数量）】,我自己设置一个【最大的迭代次数】
    def __init__(self,k,n=20):
        # 【先传值】
        self.k = k
        self.n = n
        
        # 【初始化类之后开始训练：】
    def fit(self,x,centres=None):
        
        # 【大意了，这个质心点是可以指定的，还原的不完整】
        
        
        # 【又犯这种错误！！！】
        # 怎么老是乱传东西？下面的n和k统一用self.引出！！！！
        
        # 我要【数据】
        # 首先我是不是要【随机化创建质心】？
        # x是数据：
        # 先生成质心的【索引】
        # 【对应索引的点就是我们初始化的质心】
        
        # 因为是生成索引，所以就是0-x[len]
        if centres is None:
#             idx = np.random.randint((0,len(x)),size = self.k) 哪有问题？
            idx = np.random.randint(low=0, high=len(x), size=self.k)
            # 那就正规搞
            # 下面这个是【存下质心所在索引】
            centres = x[idx]
        
#         # 创建一个【字典】
#         # 对应的键也就是对应的【质心的索引】
#         # 对应的值就是【对应的离质心最短的那些点】
#         closest_dots = {key : [] for key in centres}
#         # 下面开始进行训练：
        fit_times = 0
        
        while fit_times <=self.n:
            # 创建一个【字典】
            # 对应的键也就是对应的【质心的索引】
            # 对应的值就是【对应的离质心最短的那些点】
            closest_dots = {key : [] for key in range(self.k)}
            # 下面开始进行训练：
            
            # 【注意！！！！】因为每次更新之后都要重新选择！！！！所以是在循环里面创建数组！！！！
            
            label= []
            
            # 下面我要【找到离质心最短距离点的索引：】
            # 遍历每一个点：
            
            for every_dot in x:
                # 最简单的距离：
                # 【这个是得到最近的质心索引值！！！】
                closest_idx = np.argmin(np.sum((every_dot - centres)**2, axis = 1)**0.5)
                # 【取得到对应质心的距离是最小值的那些点】
                closest_dots[closest_idx].append(every_dot)
                label.append(closest_idx)
                
            # 最后一步：更新：
            # 对每一个质心继续求平均值：
            for i_x in range(self.k):
                x[i_x] = sum(closest_dots[i_x])/len(closest_dots[i_x])
            fit_times +=1
         
            
            
            
        return closest_dots,centres,np.array(label)
    
    
    
#     low: int
# 生成的数值最低要大于等于low。
# （hign = None时，生成的数值要在[0, low)区间内）
# high: int (可选)
# 如果使用这个值，则生成的数值在[low, high)区间。
# size: int or tuple of ints(可选)
# 输出随机数的尺寸，比如size = (m * n* k)则输出同规模即m * n* k个随机数。默认是None的，仅仅返回满足要求的单一随机数。
# dtype: dtype(可选)：
# 想要输出的格式。如int64、int等等


# 该函数最重要的参数为axis，
# np.argmin(a, axis=指定值)求的就是a在指定的第axis维度（轴）上的最小值对应的位置索引，
# 若a中在指定轴上有多个最小值，则返回第一个最小值的索引,
# axis的取值范围就是数据维数的取值范围。
# 其中a可能为list类型，也可能为array类型。
```

```python
# 重新练一下手
class MyKMeans1:
    # 传进来我们的超参数：
    def __init__(self,max_liter,k):
        self.max_liter = max_liter
        self.k = k
    def fit(self,x,centres = None):
        if centres is None:
            idx = np.random.randint(low = 0 , high = len(x), size = self.k)
            centres = x[idx]
            # 上面这步就是为了随机到我们的中心点
        fit_time = 0    
        # 下面创建一个字典：去存放我们的距离每个中心点的最近的那些点的【坐标】：
        # 这个字典在不在循环里面创建都无所谓
        closest_dots = {key : [] for key in range(self.k)}
        
        # 对了：别忘了我们的label[]
        label = []
        
        while fit_time <= self.max_liter:
            for every_dots in x:
                # 找到每一个对应的距离最近的质心
                # 这个相减返回两个值一个是横坐标的一个是纵坐标的差
                # 先得到这两个差的差的各自的平方，然后横着相加，得到距离平方
                # 再开个根号就得到距离
                # 再求距离的最近的质心的对应的索引
                closest_grav_idx = np.argmin(np.sum((every_dots - centres)**2, axis = 1)**0.5)
                
                # 再把我们的字典填上我们的对应的那些点的坐标
                closest_dots[closest_grav_idx].append(every_dots)
                
                # 填上之后
                label.append(closest_grav_idx)
                
            # 再进行我们的的坐标的更新
            for idx1 in range(self.k):
                # 为了方便：以后的x数据点数组前三个点就当作我们的新质心了
                # 以后我们就对着这x数据点数组的前三个点进行更新了
                # 第一次循环的时候就是把我们的x数据点数组的前三个点更新到我们的第一次随机到的质心的周围的数据的更新后的新数据
                x[idx1] = sum(closest_dots[idx1])/len(closest_dots[idx1])
            fit_time +=1
        return closest_dots,centres,label
```
